```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credited_card_df = pd.read_csv('/creditcard.csv')
credited_card_df.head()
```

```python
legit = credited_card_df[credited_card_df[credited_card_df.columns[-1]] == 0]
fraud = credited_card_df[credited_card_df[credited_card_df.columns[-1]] == 1]
```

```python
fraud.shape
```

```python
legit.shape
```

```python
legit_sample = legit.sample(n = 250)
credited_card_df= pd.concat([legit_sample, fraud], axis = 0)
```

```python
credited_card_df['Class'].value_counts()
```

```python
credited_card_df.groupby('Class').mean()
```

```python
X =credited_card_df.drop('Class', axis=1)
Y =credited_card_df['Class']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
ypred = model.predict(X_test)
```

```python
accuracy_score(ypred, Y_test)
```
